```python
#Ler o arquivo CSV
df = pd.read_csv('/content/drive/MyDrive/aps_failure_test_set.csv')
```

```python
#Obter dados do arquivo
print("Shape:", df.shape)
print("Primeiras Linhas: ")
print(df.head())
```

```python
#Analisar a coluna class (target/qualitativa)
print (df["class"].value_counts())
```

```python
#checar valores faltantes
print(df.isnull().sum())
```

```python
#Preenchendo dados faltantes com a média da coluna (x, preditores)
x=df.drop("class", axis=1)
y=df["class"]
```

```python
#Transformar a variavel categorica em numérica (classificação binaria)
y_binario=pd.get_dummies(y, drop_first=True).astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Dividir os dados entre conjunto de dados de treino e conjunto de dados de teste
x_train, x_test, y_train, y_test = train_test_split(x, y_binario, test_size=0.2, random_state=42, stratify = y_binario)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Padronizar/normalizar as features (preditores, x)
# Replace 'na' with np.nan
x_train = x_train.replace('na', np.nan)
x_test = x_test.replace('na', np.nan)

# Convert all columns to numeric, coercing errors
x_train = x_train.apply(pd.to_numeric, errors='coerce')
x_test = x_test.apply(pd.to_numeric, errors='coerce')


# Fill missing values with the mean of the column
x_train = x_train.fillna(x_train.mean())
x_test = x_test.fillna(x_test.mean())


scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#treinar o modelo de regressão logística
rl_model = LogisticRegression(max_iter=1000,random_state=42)
rl_model.fit(x_train_scaled, y_train)
```

```python
#Fazer predições nos dados de teste
y_pred = rl_model.predict(x_test_scaled)
```

```python
# Fazer predições sobre a probabilidade de uma observação pertencer a uma classe
y_pred_prob = rl_model.predict_proba(x_test_scaled)
```

```python
#Calcular as metricas para regresão logística (acurácia, sensibilidade, especificidade)
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

#Calcular acurácia, sensibilidade, especificidade
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
#gerar matriz de confusão
from sklearn.metrics import confusion_matrix
```

```python
#Gerar a matriz de confusão
cm = confusion_matrix(y_test, y_pred)
print (cm)
```

```python
#Mostrar as métricas calculadas
print("Acurácia:", accuracy)
print("Precisão:", precision)
print("Sensibilidade:", recall)
print("F1-Score:", f1)
```

```python
#Trainar modelo de Análise Discriminante Linear
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis #Importando pacote de analise de dados
lda_model = LinearDiscriminantAnalysis()
lda_model.fit(x_train_scaled, y_train)
```

```python
#Fazer as predições usando LDA
y_pred_lda = lda_model.predict(x_test_scaled)
```

```python
#Calcular as métricas de performance do modelo LDA (Acuracia, Sensibilidade, Especificidade)
accuracy_lda = accuracy_score(y_test, y_pred_lda)
precision_lda = precision_score(y_test, y_pred_lda)
recall_lda = recall_score(y_test, y_pred_lda)

#Mostrar as metricas calculadas
print("Acurácia (LDA):", accuracy_lda)
print("Precisão (LDA):", precision_lda)
print("Sensibilidade (LDA):", recall_lda)
```

```python
#Calcular a matriz de confusão para o modelo LDA
cm_lda = confusion_matrix(y_test, y_pred_lda)
print("Matriz de Confusão (LDA):")
print(cm_lda)
```
